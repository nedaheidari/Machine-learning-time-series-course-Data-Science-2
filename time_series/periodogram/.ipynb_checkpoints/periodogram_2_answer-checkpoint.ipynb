```python
import pandas as pd

# Load the data
df = pd.read_csv('data_planet.txt',delimiter=',' )

# Let's take a peek
print(df)
```

```python
import matplotlib.pyplot as plt

plt.errorbar(df['time'], df['observation'], yerr=df['error'], fmt='o', markersize=3)
plt.xlabel("Time [BJD]")
plt.ylabel("Radial Velocity [km/s]")
plt.title("Raw RV data")
plt.gca().invert_yaxis()  # Because RVs are often plotted with more negative values higher
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fit a linear trend: RV = a*time + b
a,c = np.polyfit(df['time'],df['observation'], deg=1)
linear_trend = a * df['time']+c

# Detrend the RVs
df['detrended'] = df['observation'] - linear_trend

# Plot
plt.figure(figsize=(10, 4))
#plt.plot(df['time'], df['observation'], label='Original RV')
#plt.plot(df['time'], linear_trend, label='Linear Trend', color='orange')
plt.plot(df['time'], df['detrended'], '.', label='Detrended', alpha=0.6)
plt.gca().invert_yaxis()
plt.legend()
plt.xlabel("Time [BJD]")
plt.ylabel("RV [km/s]")
plt.title("Linear Trend Removal")
plt.show()
```
