# Time Series Analysis
## **Introduction to Time Series Analysis**

Time series analysis is a crucial field in data science and statistics that focuses on analyzing and forecasting data collected over time. A time series is a set of observations recorded at specific time intervals, allowing researchers and analysts to study patterns, trends, and dependencies over time. Unlike traditional data analysis, which often assumes independent and identically distributed (IID) observations, time series data exhibits temporal dependencies, meaning that past values influence future observations. This unique characteristic makes time series analysis essential in a variety of fields, including finance, economics, climate science, astronomy, healthcare, and engineering.

Time series data can be **univariate** or **multivariate**. A **univariate time series** consists of observations of a single variable recorded over time, such as daily stock prices or monthly temperature readings. In contrast, a **multivariate time series** includes multiple variables recorded simultaneously, such as weather data consisting of temperature, humidity, and wind speed. Time series can also be classified into **continuous** or **discrete** datasets. **Continuous time series** data, such as temperature or river flow measurements, are recorded at every possible instant in time, while **discrete time series** data, such as quarterly sales figures or yearly population estimates, are collected at distinct and regular intervals.

Because time series data is prevalent across numerous domains, analyzing it effectively is key to making informed decisions. In finance, for example, stock market trends and economic indicators help predict market movements. In meteorology, historical temperature and precipitation data allow for more accurate weather forecasting. Similarly, astronomers rely on time series data from telescopes to study variable stars, exoplanet transits, and cosmic events. Time series analysis enables researchers to detect patterns in past observations and build models to predict future values when independent explanatory variables are unavailable.

---

## **Visualizing Time Series Data: Time Series Plots**

One of the first steps in time series analysis is data visualization, typically done using **time series plots** (or time plots). A time series plot is a two-dimensional graph where the **time component** is represented on the horizontal axis and the **measured variable** is plotted on the vertical axis. This visualization helps in identifying key characteristics of the data, such as trends, seasonality, and random variations.

There are two main types of time series based on how observations are collected:
- **Discrete time series**: Data points are recorded at fixed, separate intervals (e.g., monthly sales, quarterly GDP, or daily closing prices of stocks).
- **Continuous time series**: Data is recorded continuously over time (e.g., real-time temperature readings, ECG signals, or radiation levels in astrophysical observations).

Time plots are fundamental in understanding the nature of a time series before applying statistical or machine learning models. By examining these plots, analysts can determine whether a series follows a predictable pattern or exhibits significant randomness.

---

## **Key Characteristics of Time Series Data**

Time series data often exhibits several important features that must be understood before performing analysis:

1. **Trend**  
   - A long-term increase or decrease in the data over time.
   - Example: The gradual rise in global temperatures or the steady growth of a company’s revenue.
   - Trends can be linear or nonlinear and are often removed when analyzing seasonality and other variations.

2. **Seasonality**  
   - A pattern that repeats at fixed intervals (e.g., daily, monthly, yearly).
   - Example: Retail sales peaking during the holiday season, temperature fluctuations across seasons, or electricity demand varying between day and night.
   - Seasonal effects can often be identified by examining periodic patterns in the time plot.

3. **Cyclic Patterns**  
   - Recurrent fluctuations over longer periods, but unlike seasonality, cycles do not have fixed intervals.
   - Example: Economic recessions and expansions, stock market cycles, or sunspot cycles in astrophysics.
   - These cycles are influenced by external factors and are often harder to predict.

4. **Random Variations (Noise)**  
   - Irregular, unpredictable movements in the data that do not follow a clear trend or cycle.
   - Example: Sudden spikes in stock prices due to unexpected events or short-term fluctuations in temperature.
   - Random variations introduce uncertainty and can make forecasting challenging.

5. **Time Frame Considerations**  
   - The prediction horizon determines how far into the future a forecast can be made:
     - **Short-term forecasting** (1–2 periods ahead): Used in financial trading and short-term business operations.
     - **Medium-term forecasting** (5–10 periods ahead): Common in sales forecasting and resource planning.
     - **Long-term forecasting** (12+ periods ahead): Applied in climate projections and economic growth analysis.

Understanding these components helps analysts choose the right techniques for preprocessing and modeling time series data.

---

## **Types of Time Series Analysis**

Time series analysis can be broadly categorized into three main approaches:

1. **Descriptive Analysis**  
   - Focuses on understanding the nature of a time series by identifying patterns, seasonality, and trends.
   - Involves techniques such as decomposition (breaking down a series into trend, seasonal, and residual components).

2. **Classification Tasks**  
   - Involves categorizing time series data based on patterns.
   - Examples:
     - **Speech recognition**: Classifying spoken words based on audio waveforms.
     - **Machine failure prediction**: Identifying patterns in sensor data to classify normal vs. faulty machine behavior.

3. **Forecasting**  
   - Predicts future values based on historical data.
   - Examples:
     - **Stock market prediction**: Estimating future stock prices using past trends.
     - **Weather forecasting**: Predicting temperature, rainfall, and other climate factors.
     - **Traffic forecasting**: Estimating congestion patterns based on historical traffic data.

Forecasting is particularly challenging because many time series lack explanatory variables. Moreover, the trend, seasonality, cycles, and random variations present in the data may not always be easily or accurately identified, making model selection a critical step.

## **Time Series Preprocessing & Exploratory Data Analysis**

Before applying any predictive models to time series data, it is essential to perform thorough **preprocessing and exploratory data analysis (EDA)** to ensure data quality, detect patterns, and prepare the dataset for further analysis. Time series data is often messy, containing missing values, irregular timestamps, or noise that can distort analysis. By preprocessing and exploring the data effectively, we can better understand its structure, identify trends and seasonality, and address anomalies that may impact modeling.

One of the most common issues in time series data is **missing values**, which can arise from sensor malfunctions, data transmission errors, or inconsistent data collection intervals. Incomplete data can significantly impact forecasting models and statistical tests, making it necessary to handle missing values properly. Techniques such as **forward fill**, **backward fill**, and **linear interpolation** help estimate missing observations based on existing values. The `fillna()` method in Pandas allows us to replace missing values with a specific value, such as the mean or median of the series, while `interpolate()` provides different interpolation methods to estimate missing points more smoothly.

---
### **Interpolation**
**Interpolation** is the process of estimating missing values in a time series by using the known values around them. It helps reconstruct data points for better continuity and smoothness. Interpolation is used when:
- When there are **gaps or missing values** in the time series.
- When aligning multiple datasets with **different timestamps**.
- When upsampling (increasing the frequency) in **resampling**.

### **Types of Interpolation**
Pandas provides several interpolation methods:

#### **(1) Linear Interpolation (Default)**
This method assumes a straight-line relationship between known data points.
```python
# Apply linear interpolation
df["Interpolated"] = df["Value"].interpolate(method="linear")
```
✅ **Best for:** Small gaps with a near-linear trend.

---

#### **(2) Polynomial Interpolation**
Fits a polynomial curve through known points to estimate missing values.
```python
df["Interpolated_Poly"] = df["Value"].interpolate(method="polynomial", order=2)
```
✅ **Best for:** Curved trends, such as temperature fluctuations or stock prices.

---

#### **(3) Spline Interpolation**
Uses piecewise polynomials for a smoother fit.
```python
df["Interpolated_Spline"] = df["Value"].interpolate(method="spline", order=3)
```
✅ **Best for:** Smooth, continuous trends in scientific or economic data.

---

#### **(4) Time-Based Interpolation**
Useful when timestamps are irregular, ensuring interpolation considers **time gaps** rather than index position.
```python
df["Interpolated_Time"] = df["Value"].interpolate(method="time")
```
✅ **Best for:** Unevenly spaced time series, such as weather or financial data.

Another crucial step in preprocessing time series data is **resampling and aggregation**, which involves changing the frequency of the data to make it more suitable for analysis. For instance, financial data is often recorded at minute-level or second-level intervals, but analysts may need to **resample** it into hourly, daily, or weekly aggregates to capture broader trends. Similarly, climate and economic data may be collected on a monthly or yearly basis, but a finer temporal resolution may be required for short-term forecasting. Using the `resample()` function in Pandas, we can efficiently upsample (increase the frequency) or downsample (reduce the frequency) data by applying statistical aggregation methods such as mean, sum, median, or max. Resampling is particularly useful for handling irregular time series where observations are not recorded at fixed intervals.

### Resampling
**Resampling** involves changing the frequency of time series observations by **aggregating** or **estimating** new values. Resampling is used when:

- **Downsampling**: Convert **high-frequency** data into lower frequency (e.g., converting daily stock prices to monthly).
- **Upsampling**: Convert **low-frequency** data into higher frequency (e.g., converting monthly sales data to daily).

### **🔹 Types of Resampling**
Pandas provides powerful `.resample()` functions:

#### **(1) Downsampling: Reducing Frequency**
Aggregate data to a lower frequency using functions like `mean()`, `sum()`, `median()`.
```python
# Sample hourly time series
dates = pd.date_range(start="2024-01-01", periods=48, freq="H")
values = np.random.randint(50, 100, size=len(dates))
df = pd.DataFrame({"Date": dates, "Value": values}).set_index("Date")

# Resample to daily frequency and take the mean
df_daily = df.resample("D").mean()

print(df_daily)
```
✅ **Best for:** Summarizing high-frequency data into daily, weekly, or monthly values.

---

#### **(2) Upsampling: Increasing Frequency**
Fill missing time steps using interpolation when upsampling.
```python
# Resample from daily to hourly and interpolate missing values
df_hourly = df_daily.resample("H").interpolate(method="linear")

print(df_hourly)
```
✅ **Best for:** Creating high-resolution data from low-frequency observations.


**Time-based indexing and filtering** is another essential technique that allows us to work efficiently with time series datasets. Unlike standard datasets, which use numerical or categorical indexing, time series data is indexed by date-time values. This makes it possible to retrieve specific periods of interest, such as extracting all observations from a given year, month, or specific date range. For example, in Python’s Pandas library, we can filter time series data by selecting a particular year (`df['2023']`), a specific month (`df['2023-06']`), or a custom range (`df['2023-01-01':'2023-06-30']`). This capability simplifies data exploration and allows analysts to focus on relevant time frames for their analysis.

To analyze trends and volatility in time series data, **rolling statistics and moving averages** are commonly used. A **rolling mean** (also known as a moving average) smooths out short-term fluctuations, revealing longer-term trends by averaging data over a specified window size. Similarly, a **rolling standard deviation** helps measure variability and detect periods of increased volatility. Moving averages are widely used in financial markets to identify trends in stock prices, in climate science to monitor temperature variations, and in economic forecasting to track business cycles. In Pandas, these calculations are easily performed using the `rolling().mean()` and `rolling().std()` functions.


### **(3) Rolling Aggregation: Moving Averages**
Useful for smoothing noisy time series data by applying rolling windows.
```python
df["Rolling_Mean"] = df["Value"].rolling(window=3).mean()
```
✅ **Best for:** Trend analysis, stock price smoothing, temperature forecasting.

## **Detrending**

**Detrending** is the process of removing long-term trends from a time series to analyze its underlying fluctuations, seasonality, and cyclical components more effectively. A **trend** is the systematic increase or decrease in a time series over time, often caused by external factors like economic growth, climate change, or technological advancements.

### **Why Detrend a Time Series?**
- To **highlight short-term fluctuations** by removing long-term patterns.
- To make a time series **stationary**, which is essential for many forecasting models (e.g., ARIMA).
- To better detect **cyclic patterns** that might be hidden under an increasing or decreasing trend.
- To improve **periodicity detection** in Lomb-Scargle periodograms and spectral analysis.

---

## **2. Types of Detrending Methods**
Several techniques can be used to remove trends from time series data.

### **(1) Subtracting the Mean (Mean Detrending)**
The simplest method is to subtract the **mean value** from all data points.
$$
X_{\text{detrended}} = X_t - \bar{X}
$$
This centers the data around zero but does not account for an increasing or decreasing trend.

✅ **Best for:** Centering data but not removing systematic trends.

---

### **(2) Subtracting a Fitted Linear Trend (Linear Detrending)**
If the trend is approximately linear, it can be removed using **linear regression**.

$$
X_{\text{detrended}} = X_t - (a + bt)
$$

✅ **Best for:** Time series with a clear **linear** trend.

---

### **(3) Differencing (Removing Trend by Computing Changes)**
**Differencing** calculates the difference between consecutive time points:
\[
X_{\text{detrended}} = X_t - X_{t-1}
\]
This removes any constant trend in the data.

✅ **Best for:** Data with a **consistent trend**, making it **stationary**.

---

### **(4) Polynomial Detrending**
If the trend is **non-linear**, a polynomial function can be fitted and subtracted.

✅ **Best for:** **Curved** or **quadratic trends**, common in climate or economic data.

---

### **(5) Moving Average Detrending**
A moving average smooths the trend, and subtracting it leaves the fluctuations.

✅ **Best for:** **Short-term fluctuations**, financial data, and seasonal analysis.

---

### **(6) Hodrick-Prescott (HP) Filter**
This method decomposes a time series into **trend** and **cyclical** components.

✅ **Best for:** **Macroeconomic data** and **business cycles**.

---

## **3. Choosing the Right Detrending Method**
| **Method**              | **Best For**                                  | **Removes Trend?** |
|-------------------------|-----------------------------------------------|---------------------|
| Mean Subtraction       | Centering data but not removing trend         | ❌ No              |
| Linear Regression      | Linearly increasing or decreasing trends      | ✅ Yes             |
| Differencing           | Making a time series stationary               | ✅ Yes             |
| Polynomial Fitting     | Non-linear trends (e.g., quadratic growth)    | ✅ Yes             |
| Moving Average         | Short-term fluctuations                       | ✅ Yes             |
| HP Filter             | Business cycles & economic data               | ✅ Yes             |



````python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate synthetic time series data with a trend
time = np.arange(100)
trend = 0.5 * time  # Linear trend
noise = np.random.normal(0, 5, size=len(time))
data = trend + noise

df = pd.DataFrame({'Time': time, 'Value': data})

# Mean Detrending
df['Detrended_Mean'] = df['Value'] - df['Value'].mean()

# Plot original vs. detrended data
plt.figure(figsize=(10, 4))
plt.plot(df['Time'], df['Value'], label='Original Data')
plt.plot(df['Time'], df['Detrended_Mean'], label='Detrended (Mean)', linestyle='dashed')
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.title("Mean Detrending")
plt.show()



from scipy.stats import linregress

# Compute linear regression trend
slope, intercept, _, _, _ = linregress(df['Time'], df['Value'])

# Subtract linear trend
df['Detrended_Linear'] = df['Value'] - (slope * df['Time'] + intercept)

# Plot original vs. detrended data
plt.figure(figsize=(10, 4))
plt.plot(df['Time'], df['Value'], label='Original Data')
plt.plot(df['Time'], (slope * df['Time'] + intercept), label='Original Data')
plt.plot(df['Time'], df['Detrended_Linear'], label='Detrended (Linear)', linestyle='dashed')
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.title("Linear Detrending")
plt.show()



df['Differenced'] = df['Value'].diff()

plt.figure(figsize=(10, 4))
plt.plot(df['Time'], df['Value'], label='Original Data')
plt.plot(df['Time'], df['Differenced'], label='Differenced', linestyle='dashed')
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.title("Differencing to Remove Trend")
plt.show()



from numpy.polynomial.polynomial import Polynomial

# Fit a quadratic polynomial (degree=2)
p = Polynomial.fit(df['Time'], df['Value'], deg=2)

# Compute detrended data
df['Detrended_Poly'] = df['Value'] - p(df['Time'])

plt.figure(figsize=(10, 4))
plt.plot(df['Time'], df['Value'], label='Original Data')

plt.plot(df['Time'], df['Detrended_Poly'], label='Detrended (Polynomial)', linestyle='dashed')
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.title("Polynomial Detrending")
plt.show()


df['Moving_Avg'] = df['Value'].rolling(window=10, center=True).mean()
df['Detrended_MA'] = df['Value'] - df['Moving_Avg']

plt.figure(figsize=(10, 4))
plt.plot(df['Time'], df['Value'], label='Original Data')
plt.plot(df['Time'], df['Detrended_MA'], label='Detrended (Moving Avg)', linestyle='dashed')
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.title("Moving Average Detrending")
plt.show()


#!pip install statsmodels
from statsmodels.tsa.filters.hp_filter import hpfilter

# Apply HP Filter
cycle, trend = hpfilter(df['Value'], lamb=1600)

df['Detrended_HP'] = cycle  # Cyclical component is detrended data

plt.figure(figsize=(10, 4))
plt.plot(df['Time'], df['Value'], label='Original Data')
plt.plot(df['Time'], df['Detrended_HP'], label='Detrended (HP Filter)', linestyle='dashed')
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.title("Hodrick-Prescott Filter Detrending")
plt.show()
````

## **Folding Time Series Data to Identify Periodicity**
**Time series folding** is a technique used to detect periodic signals by aligning data points from multiple cycles onto a common phase.

### * How Does Folding Work?**
1. Choose a trial period (either from prior knowledge or by detecting it using a periodogram).
2. Compute the **phase** of each observation by normalizing the time relative to the period:
   $$
   \text{Phase} = \frac{(\text{Time} - T_0)}{\text{Period}} \mod 1
   $$
   where \( T_0 \) is a reference time (e.g., the time of first transit).
3. Sort the observations by phase instead of absolute time.
4. Plot the data with phase on the x-axis and observed values (e.g., flux) on the y-axis.
5. If the assumed period is correct, periodic features (e.g., dips in brightness) should align consistently.



## **Lomb-Scargle Periodograms**

### **1. Lomb-Scargle Periodogram**
A **Lomb-Scargle periodogram** is a mathematical tool used to identify periodic signals in unevenly spaced time series data. It is an extension of the **Fourier Transform**, designed specifically to handle datasets where measurements are not taken at regular intervals. Unlike traditional Fourier analysis, which requires evenly spaced observations, the Lomb-Scargle method enables frequency analysis for **astronomical, geophysical, financial, and biological** time series.


### **2. Why Use the Lomb-Scargle Periodogram?**
Traditional **Fourier Transform** methods work well when time series data is collected at regular time intervals, such as **hourly stock prices** or **daily temperature readings**. However, in many real-world cases:
- Data is **missing** due to interruptions.
- Observations are **irregularly spaced** (e.g., telescope observations on clear nights only).
- The time series has **gaps** that disrupt standard periodicity analysis.

The **Lomb-Scargle periodogram** solves these problems by efficiently estimating the **power spectrum** of the signal, even when the time series is unevenly sampled.

### **3. How Does the Lomb-Scargle Periodogram Work?**
The Lomb-Scargle method calculates the **power** of a given frequency $ f $ by fitting a **sinusoidal function** to the time series data and determining how well it matches the observed data.

#### **Mathematical Formulation**
Given a time series dataset with observations at times $ t_i $ and corresponding values $ y_i $, the Lomb-Scargle periodogram computes the **power spectral density** for each frequency $ f $ as:

$$
P(f) = \frac{1}{2\sigma^2} \left\{ \frac{ \left[ \sum w_i (y_i - \bar{y}) \cos \omega (t_i - \tau) \right]^2 }{\sum w_i \cos^2 \omega (t_i - \tau)} + \frac{ \left[ \sum w_i (y_i - \bar{y}) \sin \omega (t_i - \tau) \right]^2 }{\sum w_i \sin^2 \omega (t_i - \tau)} \right\}
$$

where:
- $ \omega = 2\pi f $ is the angular frequency.
- $ \tau $ is a phase shift correction to account for uneven sampling.
- $ w_i $ are weights (optional, usually set to 1 for uniform weighting).
- $ \bar{y} $ and $ \sigma^2 $ are the mean and variance of the data.

The periodogram evaluates **how much power each frequency contributes** to the signal, allowing us to determine the dominant periodic components.

---

### **4. Applications of the Lomb-Scargle Periodogram**
The method is widely used in scientific fields that rely on periodic signal detection:

#### **1. Astrophysics & Exoplanet Discovery**
- Used to **detect periodic dips in brightness** in light curves from space telescopes (e.g., **TESS, Kepler**).
- Helps identify **stellar pulsations, binary star systems, and exoplanets**.

#### **2. Climate Science & Meteorology**
- Identifies **seasonal variations** in temperature, precipitation, and atmospheric pressure.
- Used for **El Niño detection** and studying long-term climate cycles.

#### **3. Finance & Economics**
- Detects periodic trends in **stock market cycles** and **economic indicators**.
- Used for modeling repeating patterns in financial time series.

#### **4. Medicine & Biology**
- Analyzes **biological rhythms**, such as heart rate variability and circadian cycles.
- Used in **sleep cycle analysis** and medical monitoring.


````python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from astropy.timeseries import LombScargle

from astropy.table import Table
from astropy.time import Time

# Load transit data (Kepler)
#url = "http://exoplanetarchive.ipac.caltech.edu:80/data/ETSS//Kepler2/005/761/08/ktwo201505350-c01_llc_lc.tbl"
filename='ktwo201505350-c01_llc_lc.tbl'

# Read the file directly from the URL
data = Table.read(filename, format="ascii.ipac")

# Convert Kepler TIME column to actual datetime
data["TIME"] = 2454833 + data["TIME"]
data["Datetime (UTC)"] = Time(data["TIME"], format='jd').to_datetime()


# Print the first few rows
#print(data)

# Convert Astropy Table to Pandas DataFrame
df = data.to_pandas()

# Display the first few rows
#print(df.head())

#print(df.columns)

# Drop missing values
df.dropna(inplace=True)

# Plot the raw light curve
plt.figure(figsize=(10, 4))


import matplotlib.pyplot as plt

# Extract time and flux columns
time = df['Datetime (UTC)']  # Check actual column names with df.columns
flux = df['PDCSAP_FLUX']

df['Date_since_start'] = (df["Datetime (UTC)"] - df["Datetime (UTC)"].min()).dt.days

# Plot the light curve
plt.figure(figsize=(10, 4))
plt.scatter(time, flux, s=2, color="blue", alpha=0.5)
plt.xlabel("Time (days)")
plt.ylabel("Flux")
plt.title("K2-19b Light Curve")
plt.show()



# Plot the light curve
plt.figure(figsize=(10, 4))
plt.scatter(df['Date_since_start'], df['PDCSAP_FLUX'], s=2, color="blue", alpha=0.5)
plt.xlabel("Time (days)")
plt.ylabel("Flux")
plt.title("K2-19b Light Curve")
plt.show()



### **🔹 Step 2: Detect the Period using Lomb-Scargle**
#We will use the **Lomb-Scargle periodogram** to determine the best period for phase folding.

# Define frequency range
min_period = 3    # Ignore very small periods (high frequencies)
max_period = 100  # Ignore very long periods (low frequencies)
min_freq = 1 / max_period
max_freq = 1 / min_period

# Run Lomb-Scargle Periodogram with frequency range to find best period
frequencies, power = LombScargle(df["Date_since_start"], df["PDCSAP_FLUX"]).autopower()



# Run Lomb-Scargle Periodogram with frequency range to find best period
#frequencies, power = LombScargle(df["Date_since_start"], df["PDCSAP_FLUX"]).autopower(minimum_frequency=min_freq, maximum_frequency=max_freq)

best_period = 1 / frequencies[np.argmax(power[1:])]
print(f"Best Detected Period: {best_period:.5f} days")

# Plot the periodogram
plt.figure(figsize=(10, 4))
plt.plot(frequencies, power, color="b")
#plt.axvline(best_period, color="r", linestyle="--", label=f"Best Period: {best_period:.5f} days")
plt.xlabel("Frequency (1/day)")
plt.ylabel("Power")
plt.title("Lomb-Scargle Periodogram")
plt.legend()
plt.show()



### **🔹 Step 3: Phase Folding the Light Curve**
# **fold the time series** using the detected period.

# Compute phase
df["phase"] = (df["Date_since_start"] % best_period) / best_period

# Plot the folded light curve
plt.figure(figsize=(10, 4))
plt.scatter(df["phase"], df["PDCSAP_FLUX"], alpha=0.3, color='r', label="Folded Data")
plt.xlabel("Phase")
plt.ylabel("Normalized Flux")
plt.title(f"Phase-Folded Light Curve of K2-19b (P = {best_period:.5f} days)")
plt.legend()
plt.show()
#✅ **Effect:** The transit signal is now aligned at **phase = 0**, confirming periodicity.

````
