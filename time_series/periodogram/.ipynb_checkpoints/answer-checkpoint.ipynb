```python
import numpy as np
rand = np.random.default_rng(42)
t = 100 * rand.random(100)
y = np.sin(2 * np.pi * t) + 0.1 * rand.standard_normal(100)
```

```python
from astropy.timeseries import LombScargle
frequency, power = LombScargle(t, y).autopower()
```

```python
import matplotlib.pyplot as plt  
plt.plot(frequency, power)   
```

```python
from astropy.timeseries import LombScargle
import numpy as np
import matplotlib.pyplot as plt

# Given data
time = np.array([1.2, 3.5, 4.8, 7.1, 9.6, 12.2])
brightness = np.array([10.5, 10.7, 10.3, 10.8, 10.4, 10.6])


# Compute the periodogram
frequency, power = LombScargle(time, brightness).autopower()

# Convert frequency to period
period = 1 / frequency

# Plot periodogram
plt.plot(period, power)
plt.xlim(0,8)

# Find best period
best_period = period[np.argmax(power)]
print(f"The best estimated period is {best_period:.2f} days")
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from astropy.timeseries import LombScargle



# Read the file directly from the URL
data = pd.read_csv('lc_20.rdb',delimiter='\t')


ls=LombScargle(data['time'], data['flux'])

# Compute the Lomb-Scargle periodogram
frequency, power = ls.autopower(minimum_frequency=1/10,\
                                maximum_frequency=1/1)

# Convert frequency to period
period = 1 / frequency

# Plot periodogram
plt.figure(1)
plt.plot(period, power)
plt.xlim(-1,10)

# Find best period
best_period = period[np.argmax(power)]
print(f"The best estimated period is {best_period:.2f} days")


model = ls.model(data['time'], 1/best_period)
plt.figure(2,figsize=(15,3))
plt.scatter(data['time'], data['flux'],s=0.1)
plt.plot(data['time'],model,color='red')
plt.ylim(0.999,1.001)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from astropy.timeseries import LombScargle

# Create a simple synthetic dataset
time = np.linspace(0, 10, 100)  # Time from 0 to 10 years
true_period = 3  # The true period of the signal (3 years)
temperature = np.sin(2 * np.pi * time / true_period) + np.random.normal(0, 0.1, len(time))  # Sinusoidal signal + noise

# Create DataFrame
df = pd.DataFrame({'Year': time, 'Temperature': temperature})

# Task 1: Plot temperature anomalies over time
plt.figure(figsize=(10, 5))
plt.plot(df['Year'], df['Temperature'], label="Temperature Anomaly (°C)", color='blue')
plt.xlabel("Year")
plt.ylabel("Temperature Anomaly (°C)")
plt.title("Temperature Anomalies Over Time")
plt.legend()
plt.show()

# Task 2: Apply Lomb-Scargle to detect periodic trends
frequency, power = LombScargle(df['Year'], df['Temperature']).autopower()

# Convert frequency to period (T = 1/f), in **years**
period = 1 / frequency  

# Find the best period (dominant period)
best_period = period[np.argmax(power)]
print(f"The best estimated period is {best_period:.2f} years")

# Generate the best-fitting model using the best frequency (period)
best_frequency = frequency[np.argmax(power)]  # Get frequency corresponding to best period
model = LombScargle(df['Year'].to_numpy(), df['Temperature'].to_numpy()).model(df['Year'].to_numpy(), best_frequency)  # Generate model

# Task 3: Plot original data and the best-fitting model
plt.figure(figsize=(10, 5))
plt.plot(df['Year'], df['Temperature'], label="Original Data (Temperature Anomaly)", color='blue')
plt.plot(df['Year'], model, label="Best Fitting Model", color='red', linestyle='--')
plt.xlabel("Year")
plt.ylabel("Temperature Anomaly (°C)")
plt.title("Temperature Anomalies with Best Fitting Model")
plt.legend()
plt.show()
```
