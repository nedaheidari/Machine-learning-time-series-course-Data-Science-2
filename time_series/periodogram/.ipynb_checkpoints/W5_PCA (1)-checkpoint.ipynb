## Principal Component Analysis (PCA)


### Introduction
Principal Component Analysis (PCA) is a powerful multivariate analysis technique used to simplify complex datasets characterized by a large number of interrelated variables by **reducing its dimensionality**. It is a **descriptive method**, meaning all the "active" variables are considered equally important in the analysis. The primary goal of PCA is to **organize, synthesize, and reduce** information to achieve a better understanding of the underlying patterns within the data.

### **Core Objectives of PCA:**

PCA primarily aims to achieve two key objectives:

**Objective 1: Summarizing Resemblances Between Individuals/Objects (Observations)**

*   **Goal:** To understand the similarities and differences between individuals (observations) based on their characteristics described by quantitative variables. PCA allows us to study the **positions of observations with respect to each other and in terms of the variables**.
*   **Questions Addressed:**
    *   Which individuals are similar to each other in terms of their quantitative characteristics? Which are different?
    *   Can we identify **homogeneous groups** of individuals?
    *   Can we establish a **typology** for the individuals based on their characteristics?
*   **Method:** PCA achieves this by projecting the individuals from a high-dimensional space (defined by the original variables) onto a lower-dimensional space called **factor maps**. In these factor maps, **proximities between observations are interpreted as indicating similar behaviors** with respect to the variables.  Individuals close to each other on the factor map are considered to have similar profiles across the measured variables.

**Objective 2: Summarizing Relationships Between Variables**

*   **Goal:** To understand the inter-relationships between the variables themselves in the context of the observations. PCA helps to study the **relationships between the variables with regard to the observations**.
*   **Questions Addressed:**
    *   How can we **reduce the large set of related variables** to a smaller, more manageable set of synthetic variables?
    *   Can we identify a few **synthetic, uncorrelated (new) variables** to simplify the analysis and reduce dimensionality?
    *   Can we **get rid of redundancy** caused by highly correlated variables?
    *   Which variables are **positively related (correlated)** to each other? Which ones are **opposed** (negatively correlated)?
    *   Are there **groups of correlated variables** within the dataset?
*   **Method:** PCA addresses this by transforming the original set of correlated variables into a new set of **uncorrelated variables** called **principal components**. These principal components are derived in such a way that they capture the maximum variance in the data.  PCA also uses **correlation circles** to visualize the relationships between the original variables and their correlations with the principal components.

**Steps Involved in Performing PCA:**

PCA is typically carried out in a series of well-defined steps:

**Step 1: Choosing Variables and Observations (Active and Supplementary Elements)**

*   This is a crucial initial step and often involves **subjective choices**. It involves deciding which variables and observations will be considered **"active"** in the PCA computation and which will be treated as **"supplementary"**.
*   **Active Elements:** These are the variables and observations that directly drive the PCA calculations and determine the principal components. All "active" variables are treated as equally important.
*   **Supplementary Elements:**  These are variables or observations that are not directly used in the PCA computations but are projected onto the results (factor maps and correlation circles) *after* the PCA is performed on the active elements.
*   **Purpose of Supplementary Elements:**
    *   To visualize elements that are not meant to influence the core structure of the data but are of interest for interpretation.
    *   To explore how these supplementary elements relate to the structure defined by the active elements.
*   **Examples of Supplementary Elements:**
    *   **Observations:**
        *   Observations that are considered **atypical points or outliers** due to excessive contribution to the analysis.
        *   **Average observations per subgroups** that are created based on qualitative variables (e.g., average car characteristics for different car types).
    *   **Variables:**
        *   Variables that have a **particular position** compared to others and might skew the analysis focus if active.
        *   Variables that are **not directly linked to the primary themes** being studied in the dataset, but might provide additional context.

**Step 2: Standardizing the Variables**

*   **Rationale:**  Variables in a dataset often have **heterogeneous characteristics** due to different units of measurement (e.g., weight in kg, speed in km/h) or different distributions. This heterogeneity can unfairly influence PCA, where variables with larger variances might dominate the analysis.
*   **Standardization Process:** To address this, variables are **standardized** to have a mean of 0 and a standard deviation of 1.  The most common method is **Z-score standardization**, using the formula:

    ```
    z = (x - μ) / σ
    ```
    where:
    *   `z` is the standardized value.
    *   `x` is the original value.
    *   `μ` is the mean of the variable.
    *   `σ` is the standard deviation of the variable.

*   **Effect of Standardization:**
    *   **Eliminates differences in means and variances** between variables, ensuring that each variable contributes equally to the analysis, regardless of its original scale or dispersion.
    *   Results in **"Scaled PCA,"** where all variables are on a comparable scale.
*   **Checking for Extreme Observations:** After standardization, it is important to **re-check for extreme observations**. Standardization can sometimes highlight observations that become outliers in the standardized data even if they weren't initially obvious in the raw data. If extreme observations are removed, the process may need to be iterated as new extreme observations might appear.

**Step 3: Proceeding with the Analysis - Double Geometric Representation**

PCA utilizes a **double geometric representation** to analyze both individuals and variables simultaneously.

*   **Representation of Individuals:**
    *   **Conceptualization:**  Each individual, described by *p* active variables, can be visualized as a **point in a *p*-dimensional space**. Each dimension in this space corresponds to one of the *p* variables.
    *   **Scatterplot of Individuals:**  The entire dataset of *N* individuals becomes a **scatterplot of *N* points in this *p*-dimensional space**.
    *   **"Average" Individual:** The origin (O) of this space can be considered the **"average" individual**, representing the mean value across all variables for the entire dataset.
    *   **Studying Resemblance & Dispersion:** By examining this scatterplot:
        *   **Resemblance:** The **distance between two individual points** in this space represents the dissimilarity or difference between those individuals. Shorter distances indicate greater resemblance.
        *   **Dispersion:** The overall **shape of the cloud** of points and its dispersion provide insights into the variability and structure of the data.
*   **Representation of Variables:**
    *   **Conceptualization:** Each variable, measured across *N* individuals, can be visualized as a **point (or more accurately, a vector) in an *N*-dimensional space**. Each dimension in this space corresponds to one of the *N* individuals.
    *   **Scatterplot of Variables:** The set of *p* active variables becomes a **scatterplot of *p* points (vectors) in this *N*-dimensional space.**
    *   **Focus on Vectors:** In the variables' representation, we are primarily interested in the **directions of the vectors** rather than just their endpoints. This is because vector directions encode relationships.

**3.1 Representation of the Individuals - Finding Principal Components**

*   **Principal Components (PCs):** PCA aims to find a new set of orthogonal axes (dimensions) called **principal components** within the *p*-dimensional space of individuals. These PCs are derived based on the dispersion (variance) of the individual points.
*   **First Principal Component (PC1):**  PC1 is constructed to point in the **direction of maximal dispersion** of the individual points in the *p*-dimensional space. It is the direction along which the data exhibits the greatest variability. This axis passes through the "average" individual point (O).
*   **Second Principal Component (PC2):** PC2 is constructed to be **orthogonal (perpendicular)** to PC1 and is oriented in the direction of **second maximal dispersion**.  It captures the next largest amount of variance after PC1, in a direction independent of PC1. PC2 also passes through the average point (O).
*   **Subsequent Principal Components:** This process continues to find PC3, PC4, and so on, up to the *p*th principal component. Each subsequent PC is:
    *   **Orthogonal** to all preceding PCs.
    *   Oriented to capture the **maximal remaining dispersion** after accounting for the variance explained by previous PCs.
*   **Factor Maps (2D Representations):**  Since visualizing in *p*-dimensions is often impossible (especially when *p* is large), PCA uses **factor maps** to provide **optimal 2D representations**. A factor map is a **projection plane** of the *p*-dimensional individual scatterplot onto the plane defined by two principal components (typically PC1 and PC2). These 2D maps allow for easier visualization and interpretation of individual relationships.
*   **New Variables - Principal Components as Linear Combinations:**
    *   The principal components themselves are **new variables**.  There are *p* principal components, just like there were *p* original variables.
    *   **Uncorrelatedness:** A key property of PCs is that they are **uncorrelated** with each other. This eliminates redundancy present in the original set of potentially correlated variables.
    *   **Decreasing Variance:** The principal components are ordered by their **variance (inertia)**. The first principal component (PC1) has the **highest variance**, meaning it explains the largest proportion of the total variability in the data. Subsequent PCs have progressively **decreasing variances:** Var(PC1) > Var(PC2) > Var(PC3) > ... > Var(PCp). The variance of each PC (its eigenvalue or inertia) corresponds to the dispersion of the individual points along that PC's direction.
    *   **Linear Combinations:** Each principal component is a **linear combination of all the original variables**.  This means that PC1, for example, is calculated as a weighted sum of the original variables, with weights chosen to maximize the variance explained along PC1.
    *   **Factor Scores:** To determine the **position of individuals on the principal components**, we calculate **factor scores**. These scores are the coordinates of each individual projected onto each principal component. Factor scores essentially represent the value of each new principal component variable for each individual.

**3.2 Representation of the Variables - Correlation Circles**

*   **Studying Variable Relationships:**  In the variables' representation (scatterplot of variables in N-dimensional space), the goal is to understand the **linear correlation relationships** between the original variables.
*   **Geometric Representation of Correlation:** The scatterplot of variables provides a geometric interpretation of the **linear correlation coefficient**.
*   **Vector Angles & Correlation:** To summarize variable relationships, PCA studies the **angles between the vectors** representing the variables in the N-dimensional space.
*   **Property 1: Equal Distance from Origin:** All the endpoints of the vectors representing the variables are approximately at the **same distance from the origin** (approximately on a circle or hypersphere, with a radius related to the scale of the standardized variables).
*   **Property 2: Cosine of Angle = Correlation:** A fundamental property links the angle between two variable vectors and their linear correlation:

    ```
    cos(Angle between Vector X1 and Vector X2) = ρ(X1, X2)
    ```
    where `ρ(X1, X2)` is the linear correlation coefficient between variable X1 and variable X2.

    *   **Small Angle (Vectors close):** Cosine close to 1, indicating a **strong positive correlation**.
    *   **Angle around 90 degrees (Vectors orthogonal):** Cosine close to 0, indicating **weak or no linear correlation**.
    *   **Angle around 180 degrees (Vectors opposite):** Cosine close to -1, indicating a **strong negative correlation**.
*   **Need for Synthetic Variables & Optimal Projection:**  Studying angles in a high-dimensional N-space directly becomes impossible beyond 3 dimensions. Therefore, PCA constructs **optimal 2D representations called correlation circles** to visualize these relationships in a simplified manner.
*   **Correlation Circles:**
    *   **Purpose:** To provide a 2D visual representation of the **relationships between the original variables** and their correlations with the principal components.
    *   **Projection:**  A **correlation circle** is constructed by projecting the variable vectors from the N-dimensional space onto a 2D plane defined by the first two principal components (PC1 and PC2).
    *   **Interpretation:**
        *   **Vector Length:** The length of a variable vector projected onto the correlation circle indicates the **quality of representation** of that variable on the chosen PC plane. Longer vectors (closer to the circle's border) indicate better representation.
        *   **Angles between Vectors:** The angles between the projected variable vectors on the correlation circle approximate the original angles in the N-dimensional space and thus reflect the **correlations between variables**. Small angles indicate positive correlations, angles near 90 degrees indicate weak correlation, and angles near 180 degrees indicate negative correlation.
        *   **Position Relative to Axes:** The position of a variable vector relative to the PC axes (PC1 and PC2) indicates its **correlation with those principal components**. Variables pointing along the direction of PC1 are highly correlated with PC1, and similarly for PC2.

**Step 4: Studying the Principal Components**

This step involves several sub-steps focused on understanding and utilizing the derived principal components.

**4.1 Choosing the Number of Principal Components to Analyze**

*   **Goal:** To decide how many principal components to retain for further interpretation and analysis.  Typically, we aim to reduce dimensionality, so we select a subset of the total *p* PCs.
*   **Criteria for Selection:**
    *   **1. Kaiser's Rule (Eigenvalue Criterion):**
        *   **Principle:** Retain only those principal components whose **explained inertia (variance)** is **larger than the average inertia**.
        *   **Threshold:** Components with an eigenvalue **larger than 1** (in the case of standardized data) are considered to restore more information than an "average" original variable.  This is because the average eigenvalue for standardized data is 1 (sum of eigenvalues equals p, and there are p variables).
        *   **Percentage of Explained Inertia:**  Alternatively, select components whose explained inertia (as a percentage of total inertia) is larger than **(100/p)%**, where *p* is the number of active variables.
    *   **2. Scree Test (Catell's Rule - Visual Criterion):**
        *   **Principle:** Examine the **scree plot**, which is a graph of eigenvalues (or inertia) plotted against the component number (PC1, PC2, PC3...).
        *   **Identification of "Elbow":** Look for the **"elbow" or sharp bend** in the scree plot. The components *before* the elbow (those with larger eigenvalues and steeper slope) are considered important and are retained. The components *after* the elbow have smaller eigenvalues and contribute less significantly to the overall variance, suggesting diminishing returns.
        *   **Biggest Drops in Inertia:**  Focus on identifying the **largest drops in inertia (eigenvalues)** between consecutive components. Components that correspond to significant drops are often retained.
    *   **Combined Approach:** Often, both Kaiser's Rule and the Scree Test are used in combination to make a robust decision about the number of components.
    *   **Example in the Description:** The example mentions keeping the first PC "for sure" and the second PC "definitely needed" to secure at least 2 dimensions for visualization, even if the second PC might not strictly meet Kaiser's rule. This decision is confirmed by the Scree Test. However, it also cautions about potential "too much weight" being placed on the first PC if only a very small number of PCs are selected.

**4.2 Interpreting the Principal Components**

*   **Goal:** To understand what each principal component represents in terms of the original variables.
*   **Method:**  Examine the **linear correlations between the original variables and the principal components**.
*   **Interpretation based on Correlations:**
    *   **High Positive Correlation:**  If a principal component has a high positive correlation with a particular original variable, it suggests that this PC largely captures the variation associated with that variable. Individuals with high scores on this PC tend to have high values for that original variable.
    *   **High Negative Correlation:** High negative correlation means individuals with high PC scores tend to have low values for the original variable.
    *   **Low Correlation:** Low correlation suggests the PC is not strongly related to that particular original variable.
*   **Example Interpretation (from description):**
    *   **PC1 (F1):** Found to be strongly correlated with "Displacement, Power, and size (Weight/Length)" of cars.  This suggests PC1 might represent a "Car Size and Power" factor.
    *   **PC2 (F2):** Found to "separate Speed and power from Width". This indicates PC2 distinguishes between cars that prioritize speed and power versus those that emphasize width.

**4.3 Synthesizing using Correlation Circles**

*   **Correlation Circles as Synthesis Tools:** Correlation circles serve as powerful tools for synthesizing the relationships between both original variables and between variables and principal components.
*   **Visualizing Correlation Structure:** They graphically visualize the **correlation structure between the original variables**, as interpreted from the angles between variable vectors projected onto the circle.
*   **Example Interpretation (from description):**
    *   **Weight and Length:**  Vectors for "Weight" and "Length" are close together (small angle) on the correlation circle, indicating **high positive correlation** between these variables (as also seen in the correlation matrix).
    *   **Speed and Width:** Vectors for "Speed" and "Width" are somewhat close, but at a larger angle, indicating a **low positive correlation**.
    *   **No Negative Correlation Highlighted:** The example notes the absence of any pairs of vectors at near 180 degrees, meaning no strong negative correlations are visually apparent in this 2D projection.
    *   **Variable Representation Quality:**  The example also highlights that "All variables are well represented" because their vectors are "long" and "close to the circle border." This means that the 2D correlation circle effectively captures the correlations of these variables.

**Step 5: Studying the Position of the Individuals**

This final step focuses on interpreting the individual observations in the reduced dimensional space defined by the principal components.

**5.1 Evaluating Contribution to the Components**

*   **Eigenvalues and Explanatory Power:** As mentioned earlier, eigenvalues quantify the **explanatory power** of each principal component (its variance or inertia).
*   **Individual Contribution to Component Variability:** PCA calculates the **individual percentage contribution** of each observation to the variability of each principal component axis. This contribution value reflects how much each observation contributes to defining the direction and variance of that specific PC.  Observations with high contributions significantly influence the orientation and interpretation of a particular PC.

**5.2 Measuring the Quality of Representation**

*   **Proximity on Factor Maps and Similarity:**  Recall that **proximities between observations on factor maps are interpreted as similar behaviors** across the original variables.
*   **Potential for Projection Distortion:** However, it is crucial to recognize that a 2D factor map is a **projection** from a potentially higher-dimensional space. This projection can introduce **distortion**. Two individuals might appear close on the 2D factor map *not necessarily* because they were truly close in the original multi-dimensional space. Their apparent closeness could be an artifact of the projection onto a lower dimension.
*   **Squared Cosines (Quality of Representation Measure):** PCA calculates **squared cosines** for each individual with respect to each principal component (and often for the factor map plane - combination of first two PCs). Squared cosines provide a measure of **how well each object (individual) is represented** in the visualized space (e.g., the factor map plane).
    *   **High Squared Cosine:**  A squared cosine value close to 1 indicates that the individual is well-represented on that PC (or in the factor map plane). The projection is of high quality, and the individual's position on the map is a reliable reflection of its position in the original space *along those dimensions*.
    *   **Low Squared Cosine:** A squared cosine value closer to 0 indicates poor representation. The individual's position on the factor map might be misleading and not accurately reflect its true position in the original multidimensional space for those dimensions.  Interpretation of proximity for individuals with low squared cosines should be cautious.

**5.3 Analyzing the Position of the Individuals**

*   **Micro and Macro Level Analysis:** Study the position of individuals on factor maps at both:
    *   **Macro Level:** Examine the overall **structure of the cloud** of individuals. Are there clusters? What is the overall spread and orientation?
    *   **Micro Level:**  Focus on **individual positions** and their relationships to each other. Identify individuals that are close to each other, far apart, or in specific regions of the factor map.
*   **Using Factor Scores and PC Interpretations:** Analyze individual positions based on their **factor scores** (coordinates on the principal components) and the **interpretations of the principal components** (Step 4.2). For example, if PC1 is interpreted as "Car Size and Power," individuals with high positive factor scores on PC1 are interpreted as being large and powerful cars.
*   **Don't Forget Supplementary Elements:**  Crucially, remember to **project any supplementary elements (supplementary observations)** onto the factor maps. Analyze their positions in relation to the active individuals, based on the interpretations of the principal components. Similarly, **project supplementary variables** onto the correlation circles and interpret their relationships with active variables and principal components. This helps to understand the context of the supplementary elements within the structure defined by the active data.

By systematically following these steps, PCA provides a comprehensive framework for exploring, simplifying, and understanding the complex relationships within multivariate datasets, achieving the objectives of summarizing resemblances between observations and relationships between variables, and reducing data dimensionality.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


# 1. Load your data:

# Example data
# The MHEALTH (Mobile Health) dataset is devised to benchmark techniques dealing with human behavior analysis based on multimodal body sensing.    
# https://archive.ics.uci.edu/dataset/319/mhealth+dataset
#df = pd.read_csv('MHEALTHDATASET/mHealth_subject1.log', delimiter='\t', names=['col'+str(i) for i in range(24)])

df = pd.read_excel('ApplesTexturedata.xlsx')
X = df.drop("Sample Apples", axis=1).values  # Assuming "target_variable" is not needed for PCA
print(df.columns)
print(df.shape)



n_features = np.size(X.size)
#print(X)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D  # Import for 3D plotting (moved outside the function)


def perform_pca(X, n_components=2, visualize=True):
    """
    Performs Principal Component Analysis (PCA) on the given data.

    Args:
        X: The input data as a NumPy array or Pandas DataFrame.
        n_components: The number of principal components to keep.
        visualize: Whether to visualize the results (2D or 3D scatter plot).

    Returns:
        A tuple containing:
            - df_pca: A Pandas DataFrame with the principal components.
            - pca: The fitted PCA object (useful for accessing explained variance, etc.).
            - X_scaled: The scaled data used for PCA.
            - X_reconstructed: The reconstructed data (optional).
    """
    
    
    # 1. selection
    #X = X.values  # Convert DataFrame to NumPy array 
     
    n_features = X.shape #Get the number of features

    # 2. Standardize the data:
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    print(X, X_scaled)
    # 3. Apply PCA:
    pca = PCA(n_components=n_components)
    X_pca = pca.fit_transform(X_scaled)

    # 4. Explained Variance Ratio:
    explained_variance_ratio = pca.explained_variance_ratio_
    print("Explained Variance Ratio:", explained_variance_ratio)
    print("Total Explained Variance:", sum(explained_variance_ratio))

    # 5. Create DataFrame:
    df_pca = pd.DataFrame(data=X_pca, columns=[f"PC{i+1}" for i in range(n_components)])

    # 6. Visualize:
    if visualize:
        if n_components <= 2:
            plt.figure(figsize=(8, 6))
            plt.scatter(df_pca["PC1"], df_pca["PC2"])
            plt.xlabel("Principal Component 1")
            plt.ylabel("Principal Component 2")
            plt.title("PCA Results")
            plt.grid(True)
            plt.show()
        elif n_components == 3:
            fig = plt.figure(figsize=(8, 6))
            ax = fig.add_subplot(111, projection='3d')
            ax.scatter(df_pca["PC1"], df_pca["PC2"], df_pca["PC3"])
            ax.set_xlabel("Principal Component 1")
            ax.set_ylabel("Principal Component 2")
            ax.set_zlabel("Principal Component 3")
            plt.title("PCA Results (3D)")
            plt.show()
        elif n_components <= 4: #Added visualization for 4 components
            fig, axes = plt.subplots(2, 2, figsize=(10, 8)) #2x2 subplots
            pairs = [(0, 1), (0, 2), (0, 3), (1, 2)] #Combinations of PCs to plot
            for i, (pc1, pc2) in enumerate(pairs):
                row = i // 2 #Integer division to get the row index
                col = i % 2 #Modulo to get the column index
                axes[row, col].scatter(df_pca[f"PC{pc1+1}"], df_pca[f"PC{pc2+1}"])
                axes[row, col].set_xlabel(f"PC{pc1+1}")
                axes[row, col].set_ylabel(f"PC{pc2+1}")
                axes[row, col].set_title(f"PC{pc1+1} vs PC{pc2+1}")
                axes[row, col].grid(True)
            plt.tight_layout() #Adjust subplot parameters
            plt.show()

    # 8. Reconstruct (optional):
    X_reconstructed = pca.inverse_transform(X_pca)

    return df_pca, pca, X_scaled, X_reconstructed

    
````
