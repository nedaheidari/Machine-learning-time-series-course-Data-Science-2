What is the Lomb-Scargle Periodogram?

The Lomb-Scargle periodogram is a method used to detect periodic signals in data. It applies the following model to the data and tests different periods. Then, it assigns a score to each period based on how well the model fits the data. The best period is the one with the highest score.

                    y=Acos(ωt)+Bsin(ωt)+C
where:

A and B are the amplitudes of the cosine and sine components.

ω = 2πf=2π/(period)

C is a constant offset (the mean of the data).

Yes. we should first remove the trend and then apply the periodogram! 

Let's review the notebook for the trend analysis, and then we will return here.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import detrend
from statsmodels.tsa.seasonal import seasonal_decompose

df = pd.read_csv('annual.csv')

# Apply linear detrending
df["Detrended"] = detrend(df["Mean"])

# Plot the result
plt.figure(figsize=(10, 4))
plt.plot(df["Year"], df["Mean"], label="Original Data", color="black", alpha=0.5)
plt.plot(df["Year"], df["Detrended"], label="Detrended (SciPy)", color="green")
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Trend Removal using Linear Detrending")
plt.legend()
plt.show()
```

```python
# Compute moving average (window size = 10)
df["Moving_Avg"] = df["Mean"].rolling(window=10, center=False).mean()

# Subtract moving average from original data
df["Moving_Detrended1"] = df["Mean"] - df["Moving_Avg"]


# Plot the result
plt.figure(figsize=(10, 4))
plt.plot(df["Year"], df["Mean"], label="Original Data", color="black", alpha=0.5)
plt.plot(df["Year"], df["Moving_Detrended1"], label="Detrended (Moving Average)", color="red")
plt.plot(df["Year"], df["Moving_Avg"], label="trend", color="green")
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Trend Removal using Moving Average")
plt.legend()
plt.show()
```

```python
# Perform quadratic fit (degree 2 polynomial)
a, b, c = np.polyfit(df["Year"], df["Mean"], 2)  # Degree 2 for quadratic fit


y_fit = a * df["Year"]**2 + b * df["Year"] + c  # Quadratic model values

df["Moving_Detrended1"]=df["Mean"]- y_fit

# Plot the result
plt.figure(figsize=(10, 4))
plt.plot(df["Year"], df["Mean"], label="Original Data", color="black", alpha=0.5)
plt.plot(df["Year"], df["Moving_Detrended1"], label="Detrended (Moving Average)", color="red")
plt.plot(df["Year"], y_fit, label="trend", color="green")
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Trend Removal using Moving Average")
plt.legend()
plt.show()
```

```python
# Apply Lomb-Scargle to the detrended data
frequency, power = LombScargle(df['Year'], df["Moving_Detrended1"]).autopower()
ls=LombScargle(df['Year'], df["Moving_Detrended1"]) 
frequency, power =ls.autopower()

# Convert frequency to period (T = 1/f), in **years**
period = 1 / frequency  

# Plot the periodogram
plt.figure(figsize=(8, 5))
plt.plot(period, power, color='red')
plt.xlim(0,100)
plt.xlabel("Period (years)")
plt.ylabel("Power")
plt.title("Lomb-Scargle Periodogram of Detrended Global Temperatures")
plt.gca().invert_xaxis()  # Shorter periods on the right
plt.show()

# Find the best period (dominant period)
best_period = period[np.argmax(power)]
print(f"The best estimated period is {best_period:.2f} years")
```

```python
model = ls.model(df['Year'],1/best_period)
plt.figure(2,figsize=(15,3))
plt.scatter(df['Year'], df["Moving_Detrended1"],s=0.8)
plt.plot(df['Year'],model,color='red')
```
