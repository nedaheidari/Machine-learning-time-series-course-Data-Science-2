# üß† 1. Quick Recap of DBSCAN 

DBSCAN = Density-Based Spatial Clustering of Applications with Noise

Main parameters:

eps: max distance between neighbors

min_samples: min points to form a dense region (2* number of features)

Strengths: finds arbitrarily shaped clusters, identifies noise

# Challenge: how to choose eps? 

‚Üí We use the elbow method

üìê What is the elbow method for DBSCAN? 

1. For each point, we compute the distance to its k-nearest neighbor

2. Sort those distances

3. Plot ‚Üí look for an elbow (sharp bend = good choice for eps)

Where to bend:

Before the distances start rising quickly ‚Üí separates clusters from noise

# Code : using the elbow method with DBSCAN 

Step-by-step:

1. import the library (from sklearn.neighbors import NearestNeighbors)
   
3. Choose the value of k (k = min_samples).

4. Initialize the nearest neighbor search (nn = NearestNeighbors(n_neighbors=k)). This sets up a model that can find the k nearest neighbors for any point in our dataset.

5. Fit it to our data (nn.fit(X)). Now the model knows where all the data points are.

6. Find k-nearest neighbors for every point. This finds the distances from each point to its k nearest neighbors (distances, _ = nn.kneighbors(X)).

7. Take only the k-th distance for each point and sort them (k_distances = np.sort(distances[:, -1]))

```python
k = 4 

nn = NearestNeighbors(n_neighbors=k)
nn.fit(X)
distances, _ = nn.kneighbors(X)

# Take the distance to the k-th neighbor
k_distances = np.sort(distances[:, -1])

# Step 3: Plot the k-distance graph
plt.plot(k_distances)
plt.ylabel(f"Distance to {k}th nearest neighbor")
plt.title("Elbow Method for Choosing eps")
plt.grid()
plt.show()
```
