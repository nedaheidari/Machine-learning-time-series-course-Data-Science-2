## Clustering Methods: Unveiling Patterns Through Grouping

### Introduction:

Clustering methods are a powerful set of techniques employed to group individuals or objects into homogeneous clusters based on their inherent similarities. This process is crucial for discovering hidden patterns and structures within data. Often, clustering follows dimensionality reduction techniques like Correspondence Analysis (CA), Principal Component Analysis (PCA), or Multiple Correspondence Analysis (MCA), which simplify the data and enhance the effectiveness of clustering.
Real-World Applications could be customer segmentation, gene expression analysis, astronomy (grouping similar stars/galaxies), anomaly detection.



### Distance Metrics

- Euclidean Distance  
   $d(\mathbf{x}, \mathbf{y}) = \sqrt{\sum_{i=1}^{n} (x_i - y_i)^2}$
    
   - **Interpretation**: Geometric (straight-line) distance in n-dimensional space.  
   - **Usage**: Default in many algorithms (K-means, hierarchical with Ward linkage, etc.).  
   - **Effect on Clusters**: Tends to produce spherical/clumpy clusters.

2. **Manhattan Distance**  
   $d(\mathbf{x}, \mathbf{y}) = \sum_{i=1}^{n} \lvert x_i - y_i \rvert$  
   - **Interpretation**: Sum of absolute differences along each dimension.  
   - **Usage**: Effective when data has an underlying grid-like structure or strong coordinate-based features.  
   - **Effect on Clusters**: Leads to diamond-shaped clusters in 2D; handles outliers more robustly than Euclidean in some cases.

3. **Cosine Similarity**  
   $\text{similarity}(\mathbf{x}, \mathbf{y}) = \frac{\mathbf{x} \cdot \mathbf{y}}{\|\mathbf{x}\| \|\mathbf{y}\|}$  
   - **Interpretation**: Measures the angle between vectors. Often used for text data, high-dimensional data, or any scenario where **direction** matters more than magnitude.  
   - **Effect on Clusters**: Items with smaller angles between them (similar directions) cluster together, even if magnitudes differ.

4. **Effects of Choosing Different Metrics**  
   - **Cluster Shape**:  
     - Euclidean → Spherical clusters  
     - Manhattan → Diamond-shaped or block-like clusters  
     - Cosine → Clusters formed by direction rather than magnitude  
   - **Computation**: Cosine similarity can be more efficient for sparse vectors (typical in text mining).  
   - **Practical Advice**: Evaluate domain and data structure to pick the appropriate metric.

For example, if your data is text-based, you might consider using cosine similarity, which measures the angle between document vectors in a high-dimensional space.
If your data is gene expression, you might use euclidean distance or correlation based distances, to see how the expression of genes corelates.
If your data is spatial coordinates, euclidean distance is a common choice, but other metrics like Manhattan distance or geodesic distance might be more appropriate depending on the context.

### Scalability & Dimensionality
The challenges of clustering when data is high-dimensional or very large.

1. **High-Dimensional Spaces**  
   - **Curse of Dimensionality**: Distances become less meaningful; points become “equidistant” as dimensionality grows.  
   - **Sparsity**: In high-dimensional data (e.g., text, genomics), many dimensions are zero or near-zero for each data point.  
   - **Dimensionality Reduction**: PCA or t-SNE can help visualize and cluster high-dimensional data more effectively.

2. **Scalability with Large Datasets**  
   - **K-means**: Usually $O(n \cdot k \cdot d)$ per iteration (n=points, k=clusters, d=dimensions).  
   - **Hierarchical Clustering**: $O(n^2)$ or worse, so can be problematic for very large n.  
   - **Mini-Batch K-means** or approximate methods may be needed for extremely large data.

3. **Data Preprocessing**  
   - **Normalization** (e.g., Min-Max, StandardScaler) ensures each dimension contributes equally to distance.  
   - **Categorical Data**: Need specialized encodings or distance metrics (e.g., Hamming distance).

Selecting an appropriate clustering algorithm is crucial for achieving meaningful results. Clustering methods can be broadly categorized into:

* **Non-Hierarchical Methods:**
* **Mixed Methods:**

**2.1 Aggregation Criterion**

* **Grouping Strategy:** Clustering algorithms group data points based on a similarity or dissimilarity measure.
* **Similarity/Dissimilarity Measure:** In geometric representations, distance metrics are commonly used.
* **Total Inertia:** This is a measure of the overall dispersion of data points.
    * **Intercluster Inertia:** Represents the dispersion between cluster centroids.
    * **Intracluster Inertia:** Represents the dispersion within each cluster.
    * **Relationship:** Total Inertia = Intercluster Inertia + Intracluster Inertia.

**2.2 Partitioning Quality**

* **Clustering Quality:** This metric, often expressed as the ratio of Intercluster Inertia to Total Inertia, indicates the percentage of data variation explained by the clustering. A higher value indicates better clustering quality.

**2.3 Non-Hierarchical Methods**

* **Advantages:**
    * Simple and efficient, relying on iterative processes.
    * Applicable to large datasets.
* **Disadvantages:**
    * Requires specifying the number of clusters in advance.
* **K-Means Method:**
    * **Example:** Forming two clusters:
        1.  Choose two initial data points as cluster centers.
        2.  Assign data points to the nearest cluster center.
        3.  Calculate the centroids (mean) of the resulting clusters.
        4.  Reassign data points to the nearest new centroids.
        5.  Iterate until cluster assignments stabilize.

**2.4 Hierarchical Methods (Ascending)**

* **Advantages:**
    * Creates a hierarchical structure (dendrogram) that allows for the determination of the number of clusters a posteriori.
    * Simple and intuitive, involving successive merging of smaller groups.
* **Disadvantages:**
    * Not suitable for very large datasets.
* **Dendrogram Example:**
    * **Ward's Aggregation Criterion:** At each step, merges the two clusters that minimize the increase in intracluster variance.
    * **Choosing the Partition:**
        * Observe the dendrogram's overall shape.
        * Identify significant jumps in the merging distances.
        * Aim for a reasonable number of clusters that are interpretable.

**2.5 Mixed Methods**

* Combines the strengths of hierarchical and non-hierarchical methods for consolidated clustering.
* **Example:**
    * Step 1: Perform hierarchical clustering using Ward's distance.
    * Step 2: Refine the clusters using K-means.

**Step 3: Applying a Mixed Clustering Method**

As described in 2.5, here is where the combined steps are performed.

**Step 4: Characterizing the Clusters**

Three approaches are used to describe and interpret the resulting clusters:

**4.1 Identification Using Prototypes**

* **Actual Cluster Prototype (Medoid):** The most representative data point within a cluster, closest to the cluster's centroid.

**4.2 Identification Using Principal Components or Axes**

* Analyze the values associated with the principal components or axes to understand the characteristics of each cluster in the reduced dimensional space.
* Establish a cluster profile based on these values.

**4.3 Identification Using Original Variables and/or Modalities**

* Perform statistical tests (e.g., t-tests, chi-square tests) to compare the mean or frequency of variables/modalities within each cluster to the overall mean or frequency.
* Develop a cluster profile based on the significant differences observed.



### Practical steps

### **3. Choosing the Number of Clusters **
**Objective**: Explore practical methods to determine the optimal **k** for algorithms like K-means or to select a cutoff in hierarchical clustering.

1. **Elbow Method**  
   - **Idea**: Compute within-cluster sum of squares (WCSS) for different k. Look for a “bend” or “elbow” in the plot.  
   - **Practical**: Not always a clear elbow; interpret results with caution.  


2. **Silhouette Coefficient**  
   - Measures how similar a point is to its own cluster vs. other clusters.  
   - Range: \([-1, 1]\). Higher values → better-defined clusters.  

   - **Interpretation**: A silhouette near 1 means points are well-clustered, near 0 means overlapping clusters, and negative means misassigned points.

3. **Gap Statistic**  
   - Compares within-cluster dispersion with an expected dispersion under a null reference distribution.  
   - Less commonly implemented but can be powerful.

4. **Cutting a Dendrogram (Hierarchical)**  
   - You decide on a distance threshold or a desired cluster count.  
   - Visual inspection of the dendrogram to see stable merges.


## Load the dataset

We will start by loading the digits dataset. This dataset contains handwritten digits from 0 to 9. In the context of clustering, one would like to group images such that the handwritten digits on the image are the same.

```python
import numpy as np

from sklearn.datasets import load_digits

data, labels = load_digits(return_X_y=True)
(n_samples, n_features), n_digits = data.shape, np.unique(labels).size

print(f"# digits: {n_digits}; # samples: {n_samples}; # features {n_features}")
```

## Define our evaluation benchmark

We will first our evaluation benchmark. During this benchmark, we intend to compare different initialization methods for KMeans. Our benchmark will:

create a pipeline which will scale the data using a StandardScaler;

train and time the pipeline fitting;

measure the performance of the clustering obtained via different metrics.

```python
from time import time

from sklearn import metrics
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler


def bench_k_means(kmeans, name, data, labels):
    """Benchmark to evaluate the KMeans initialization methods.

    Parameters
    ----------
    kmeans : KMeans instance
        A :class:`~sklearn.cluster.KMeans` instance with the initialization
        already set.
    name : str
        Name given to the strategy. It will be used to show the results in a
        table.
    data : ndarray of shape (n_samples, n_features)
        The data to cluster.
    labels : ndarray of shape (n_samples,)
        The labels used to compute the clustering metrics which requires some
        supervision.
    """
    t0 = time()
    estimator = make_pipeline(StandardScaler(), kmeans).fit(data)
    fit_time = time() - t0
    results = [name, fit_time, estimator[-1].inertia_]

    # Define the metrics which require only the true labels and estimator
    # labels
    clustering_metrics = [
        metrics.homogeneity_score,
        metrics.completeness_score,
        metrics.v_measure_score,
        metrics.adjusted_rand_score,
        metrics.adjusted_mutual_info_score,
    ]
    results += [m(labels, estimator[-1].labels_) for m in clustering_metrics]

    # The silhouette score requires the full dataset
    results += [
        metrics.silhouette_score(
            data,
            estimator[-1].labels_,
            metric="euclidean",
            sample_size=300,
        )
    ]

    # Show the results
    formatter_result = (
        "{:9s}\t{:.3f}s\t{:.0f}\t{:.3f}\t{:.3f}\t{:.3f}\t{:.3f}\t{:.3f}\t{:.3f}"
    )
    print(formatter_result.format(*results))
```

## Run the benchmark
We will compare three approaches:

an initialization using k-means++. This method is stochastic and we will run the initialization 4 times;

a random initialization. This method is stochastic as well and we will run the initialization 4 times;

an initialization based on a PCA projection. Indeed, we will use the components of the PCA to initialize KMeans. This method is deterministic and a single initialization suffice.

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

print(82 * "_")
print("init\t\ttime\tinertia\thomo\tcompl\tv-meas\tARI\tAMI\tsilhouette")

kmeans = KMeans(init="k-means++", n_clusters=n_digits, n_init=4, random_state=0)
bench_k_means(kmeans=kmeans, name="k-means++", data=data, labels=labels)

kmeans = KMeans(init="random", n_clusters=n_digits, n_init=4, random_state=0)
bench_k_means(kmeans=kmeans, name="random", data=data, labels=labels)

pca = PCA(n_components=n_digits).fit(data)
kmeans = KMeans(init=pca.components_, n_clusters=n_digits, n_init=1)
bench_k_means(kmeans=kmeans, name="PCA-based", data=data, labels=labels)

print(82 * "_")
```

## Visualize the results on PCA-reduced data

PCA allows to project the data from the original 64-dimensional space into a lower dimensional space. Subsequently, we can use PCA to project into a 2-dimensional space and plot the data and the clusters in this new space.

```python
import matplotlib.pyplot as plt

reduced_data = PCA(n_components=2).fit_transform(data)
kmeans = KMeans(init="k-means++", n_clusters=n_digits, n_init=4)
kmeans.fit(reduced_data)

# Step size of the mesh. Decrease to increase the quality of the VQ.
h = 0.02  # point in the mesh [x_min, x_max]x[y_min, y_max].

# Plot the decision boundary. For that, we will assign a color to each
x_min, x_max = reduced_data[:, 0].min() - 1, reduced_data[:, 0].max() + 1
y_min, y_max = reduced_data[:, 1].min() - 1, reduced_data[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Obtain labels for each point in mesh. Use last trained model.
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1)
plt.clf()
plt.imshow(
    Z,
    interpolation="nearest",
    extent=(xx.min(), xx.max(), yy.min(), yy.max()),
    cmap=plt.cm.Paired,
    aspect="auto",
    origin="lower",
)

plt.plot(reduced_data[:, 0], reduced_data[:, 1], "k.", markersize=2)
# Plot the centroids as a white X
centroids = kmeans.cluster_centers_
plt.scatter(
    centroids[:, 0],
    centroids[:, 1],
    marker="x",
    s=169,
    linewidths=3,
    color="w",
    zorder=10,
)
plt.title(
    "K-means clustering on the digits dataset (PCA-reduced data)\n"
    "Centroids are marked with white cross"
)
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Generate sample data
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
```

```python
# Perform agglomerative clustering
agg_clustering = AgglomerativeClustering(n_clusters=4)
y_pred = agg_clustering.fit_predict(X)
```

```python
# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='rainbow')
plt.title('Agglomerative Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
# Generate the linkage matrix
Z = linkage(X, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```
